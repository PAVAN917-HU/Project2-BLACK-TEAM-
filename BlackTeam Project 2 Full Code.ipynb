```python
connection = sqlite3.connect('bp.db') #creating database
 

cursor = connection.cursor() #connecting database
```

Allowing Python code to execute PostgreSQL command in a database session</h4>

```python
path1.to_sql('bp1',connection) # add the table
path2.to_sql('bp2',connection) #add the table

connection.commit()
cur = connection.cursor() #connecting database
```

```python
create_table = '''CREATE TABLE bp3(
                STATIONNAME TEXT,
                COUNTYCODE TEXT,
                Latitude REAL,
                Longitude REAL,
                Date TEXT,
                STATIONID TEXT,
                WaterElevationftabovesealevel REAL,
                WaterLevelftbelowlandsurface REAL,
                StateWellNumber INTEGER,
                YEAR INTEGER,
                EllipsoidHeightmeters REAL);'''

# Creating the table into our
# database
cursor.execute(create_table)
```

```python
cursor = connection.cursor()
sql = '''INSERT INTO "bp3" SELECT bp1.STATIONNAME,bp1.COUNTYCODE,bp1.Latitude,bp1.Longitude,bp1.Date,bp2.STATIONID,bp1.WaterElevationftabovesealevel,bp1.WaterLevelftbelowlandsurface,bp1.StateWellNumber,bp2.YEAR,bp2.EllipsoidHeightmeters
FROM bp1
LEFT JOIN bp2
USING (STATIONID);'''
  
# Executing the query
cursor.execute(sql)
```
